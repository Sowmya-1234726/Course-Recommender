```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/Users/ramyaramzz/Desktop/CourseRecommendedRa/faculty_dataset.csv')

# Preprocess: One-hot encode categorical columns
df_encoded = pd.get_dummies(df.drop(columns=['RecommendedTool']))
```

```python
from sklearn.neighbors import NearestNeighbors

# Build KNN model
model = NearestNeighbors(n_neighbors=3, metric='cosine')
model.fit(df_encoded)

# Test: Pick a sample faculty to find similar ones
sample_input = df_encoded.iloc[10].values.reshape(1, -1)
distances, indices = model.kneighbors(sample_input)

print(indices)  # Shows index numbers of similar faculty
```

```python
import matplotlib.pyplot as plt

df['RecommendedTool'].value_counts().plot(kind='bar')
plt.title('Most Recommended Tools')
plt.show()
```
